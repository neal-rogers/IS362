```python
# Import libraries and modules
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
```

```python
#Read the airport CSV file
floc1 = r'D:\Projects\Educational\IS362\week-5\data\airports.csv'
ap_csv = pd.read_csv(floc1)
ap_csv
```

```python
# Create a copy of the airports DataFrame
airports_cp = ap_csv.copy()
```

```python
# Clean up the data be extracting only airport records from the CSV 
airports = airports_cp[(airports_cp.name.str.contains("Station") == False) & (airports_cp.name.str.contains("Rail") == False)]
airports
```

```python
# Extract only airports in GMT -5 (tz)and without invalid tzone values.
extract1 = airports[(airports.tz == -5.0) & (airports.tzone <> '\N')]
extract1
```

```python
# Extract airports in GMT -9 (tz), without invalid tzone values, and aren't railroad stations.
extract2 = airports[(airports.tz == -9.0) & (airports.tzone <> '\N') & (airports.name.str.contains("Station") == False) & (airports.name.str.contains("Rail") == False)]
extract2
```

```python
# Sort the airports by longitude and return the first row.
lon_sorted = extract1.sort_values(by='lon', ascending=False)
lon_sorted.head(1)
```

```python
# Sort the airports by latitude and return the first row.
lat_sorted = extract2.sort_values(by='lat', ascending=False)
lat_sorted.head(1)
```
